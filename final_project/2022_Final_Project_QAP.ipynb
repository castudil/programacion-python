## Introducción

El objetivo del problema de asignación cuadrática (QAP, por sus siglas en inglés) consiste en asignar en mis plantas en ubicaciones de tal manera de minimizar el costo de asignación. El costo de asignación se percibe como la suma sobre todos los pares, del flujo entre pares de instalaciones, multiplicado por las distancias que existen entre las ubicaciones de estas plantas.

## Ejemplo

Considere el problema de ubicación 4 de plantas (Y consecuentemente cuatro ubicaciones). Una asignación posible se muestra en la figura de más abajo: la planta dos se asigna la ubicación, la planta uno es asignada la ubicación dos, la planta cuatro se asigna a la ubicación tres, y la planta 3 es asignada a la ubicación cuatro.

qap.png￼

Esta solución podría expresarse a través de una permutación p={2,1,4,3}, qué representa que en la ubicación uno se encuentra la planta dos, en la ubicación dos se ubica la planta uno, en la ubicación tres la planta cuatro, y finalmente en la ubicación cuatro en la planta tres.

Para poder calcular el costo de asignación, es necesario conocer el flujo que existe entre las plantas. A mayor valor significa que el flujo es más intenso. EL grosor de las líneas que comunican las plantas representa la cantidad de flujo entre ellas, entre más gruesa la conexión, existe un mayor flujo entre las plantas respectivas.

El flujo entre las plantas se cuantifica, y en el caso de este ejemplo el flujo es el siguiente:

Para poder calcular el costo, también necesitamos conocer las distancias que existen entre las plantas

La información del flujo y las distancias las estamos presentando aquí a través de tablas, pero cuando se tienen más plantas y ubicaciones, entonces se hace más eficiente almacenar la información en matrices. Por ejemplo las matriz D donde el elemento D_{i,j} indica la distancia que existe entre las ubicaciones i y j.

Para poder calcular el costo total asociado a una ubicación de las plantas, entonces multiplicamos el flujo que se produce entre dos plantas por la distancia que la separa. Es necesario sumar todos los términos obtenidos para todos los pares de plantas.

En el ejemplo de más arriba se obtendría que el costo total asociado es el siguiente 

  c(p) =

	= f(1,2)⋅d(2,1)+f(1,4)⋅d(2,3)+f(2,4)⋅d(1,3)+f(3,4)⋅d(3,4)

	= 3⋅22+2⋅40+1⋅53+4⋅55

	= 419

Es necesario destacar que este costo no es el óptimo para el problema en particular.


## Formulación matemática

$N=\{1,2,\ldots,n\}$ es el número de plantas, que coincide con el número de ubicaciones

$S_n=p:N\rightarrow N$ Es el conjunto de todas las permutaciones posibles


# Parámetros

$F=(f_{i,j})$ es una matriz de $n \times n$ dónde $f_{i,j}$ corresponde al flujo entre las plantas i y j.
$D=(d_{i,j})$ es una matriz de $n \times n$ dónde $d_{i,j}$ representa a la distancia entre las ubicaciones i y  j.

Entonces el problema de optimización es el siguiente:

$min_{p \in S_n}, \sum_{i=1}^{n}  \sum_{j=1}^{n}  f_{i,j} \times d_{p_i,p_j}$

La asignación de las plantas a ubicaciones se representa por una permutación p, dónde p_i corresponde a la ubicación en donde la planta i se asigna. Cada producto individual  f_{i,j} x d_{p_i,p_j} corresponde al costo de asignar la planta i a la ubicación p_i y la planta j a la ubicación p_j.
